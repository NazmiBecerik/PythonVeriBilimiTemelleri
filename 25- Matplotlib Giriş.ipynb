```python
import numpy as np ; import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi=np.random.randint(10,70,8)
yasListesi
```

```python
kiloListesi=np.random.randint(1,120,8)
kiloListesi
```

```python
plt.plot(yasListesi,kiloListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo-Yaş Grafiği")
```

```python
numpyDizisi2=numpyDizisi1**3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r")
```
