```python
benimAdim="Nazmi"
#Global
def benimFonksiyonum():
    benimAdim="Nazmi1"
    #Enclosing
    def icFonksiyon():
        benimAdim="Nazmi2"
        #Local
        print(benimAdim)
    icFonksiyon()
```

```python
benimFonksiyonum()
```

```python
print(benimAdim) 
#Globalde çağırmış oldum
```
