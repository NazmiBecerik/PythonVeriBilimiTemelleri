## args & kwargs (argümanlar ve anahtar kelime argümanları)

```python
def yeniToplama(*args):
    return sum(args)
```

```python
yeniToplama(100,200,300)
```

```python
 yeniToplama(10,20)
```

```python
def yeniFonksiyon(**kwargs):
    return(kwargs)
```

```python
yeniFonksiyon(elma=100,armut=200,ananas=300)
```

```python
type(yeniFonksiyon(elma=100,armut=200,ananas=300))
```

```python
def keyWordKontrolu(**kwargs):
    if "nazmi" in kwargs:
        print("nazmi var")
    else:
        print("nazmi yok")
```

```python
keyWordKontrolu(esra=22,feyza=33)
```
