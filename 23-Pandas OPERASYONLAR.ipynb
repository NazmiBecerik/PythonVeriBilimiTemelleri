## group by

```python
maasSozlugu={"Departmanlar":["YAZILIM","YAZILIM","PAZARLAMA","PAZARLAMA","HUKUK","HUKUK"],
            "CalisanIsmi:":["Ahmet","Mehmet","Burak","Zeynep","Atıl","Fatma"],
             "Maas":[100,150,200,250,300,400]}
```

```python
maasDataFrame=pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
grup=maasDataFrame.groupby("Departmanlar")
```

```python
grup
```

```python

grup.mean()
```

```python
grup.count()
```
