```python
def hesapMakinesi(a,b,islem):
    if islem not in "+-*/":
        print("Lütfen belirtilen işlemlerden birini seçiniz")
    if islem == "+":
        return (str(a)+"+"+str(b)  +"=" +str(a+b))
    if islem == "-":
        return (str(a)+"-"+str(b) +"="+ str(a-b))
    if islem == "*":
        return (str(a)+"*"+str(b) +"="+ str(a*b))
    if islem == "/":
        return (str(a)+"/"+str(b) +"="+ str(a/b))
    
while True :
    try:
        a=int(input("ilk sayi:"))
        b=int(input("ikinci sayi:"))
        islem=(input("klavyeden islem seciniz:"))
        print(hesapMakinesi(a,b,islem))
    except:
        print("Tanımlanamayan değişken girdiniz , lütfen yeniden deneyiniz")
```
