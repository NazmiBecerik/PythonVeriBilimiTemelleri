## İnheritiance (kalıtım , miras bırakma )

```python
class hayvan ():
    def __init__ (self):
        print("hayvan sınıfı initi çağırıldı")
    def method1 (self):
        print("hayvan sınıfı method1 çağırıldı")
    def method2 (self):
        print("hayvan sınıfı method2 çağırıldı")
```

```python
benimHayvanım=hayvan()
```

```python
benimHayvanım.method1()
```

```python
benimHayvanım.method2()
```

```python
class kedi (hayvan):
    def hayvan__init__(self):
        hayvan.__init__(self)
        print("kedi için özel init çağırıldı")
```

```python
benimKedi=kedi()
```

```python
benimKedi.method1()
```

```python
## override yapmayı öğrendik en son hayvanda ki method 1 çağrılmasın istiyorsak kedi sınıfına yeni bir method1 oluşturalım.
```

## Polymorphism

```python
class Elma():
    def __init__(self,isim):
        self.isim=isim
    def bilgiVer(self):
        return self.isim + " 100 kaloridir "
```

```python
class Muz():
    def __init__(self,isim):
        self.isim=isim
    def bilgiVer(self):
        return self.isim + " 150 kaloridir "
```

```python
elma=Elma("elma")
```

```python
elma.bilgiVer()
```

```python
muz=Muz("muz")
```

```python
muz.bilgiVer()
```

```python
meyveListesi=[elma,muz]
```

```python
for meyve in meyveListesi:
    print(meyve.bilgiVer())
```

```python
def bilgiAl(meyve):
    print(meyve.bilgiVer())
```

```python
bilgiAl(muz)
```

```python
bilgiAl(elma)
```
