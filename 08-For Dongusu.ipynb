```python
# if döngüsünde kullandığımız işlem algoritmaları tek değişken üzerinde yaparken for döngüsünde kaç tane varsa örneğin
# 1000 tane 500 tane varsa hepsini kullanabiliyoruz.
```

```python
print("Döngü Başladı")
for numara in benimListem:
    print(numara)
print("Döngü Bitti")
```

```python
for num in benimListem:
    print(num*5/3)
print("Döngü Tamamlandı")
```
