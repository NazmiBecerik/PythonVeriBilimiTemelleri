## TRY & EXCEPT & ELSE & FİNALLY 

```python
while True:
    try:
        int(input("Lütfen bir sayı giriniz:"))
    except:
        print("Gerçek bir sayı giriniz")
        continue
    else:
        print("Teşekkürler")
    finally:
        print("Finally her türlü çıktı alır .")
```
